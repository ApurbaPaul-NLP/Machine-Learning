```python
import tensorflow as tf
import numpy as np
import random
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```python
celsius=random.sample(range(-40, 100), 100)
fahrenheit=[]
for i in range(len(celsius)):
  x=celsius[i]*1.8+32
  fahrenheit.append(x)
print(celsius)
print(fahrenheit)
```

```python
celsius_q=np.array(celsius,dtype=float)
print(celsius_q)
fahrenheit=[int(a) for a in fahrenheit]  #float to integer
fahrenheit_a=np.array(fahrenheit,dtype=float)
print(fahrenheit_a)
```

# **Create the model**

```python
layer1=tf.keras.layers.Dense(units=1,input_shape=[1])
layer2=tf.keras.layers.Dense(units=4)
layer3=tf.keras.layers.Dense(units=8)
layer4=tf.keras.layers.Dense(units=16)
layer5=tf.keras.layers.Dense(units=32)
layer6=tf.keras.layers.Dense(units=64)
layer7=tf.keras.layers.Dense(units=128)
layer8=tf.keras.layers.Dense(units=256)
layer9=tf.keras.layers.Dense(units=128)
layer10=tf.keras.layers.Dense(units=64)
layer11=tf.keras.layers.Dense(units=32)
layer12=tf.keras.layers.Dense(units=16)
layer13=tf.keras.layers.Dense(units=8)
layer14=tf.keras.layers.Dense(units=4)
layer15=tf.keras.layers.Dense(units=1)
```

```python
model = tf.keras.Sequential([layer1,layer2,layer3,layer4,layer5,layer6,layer7,layer8,layer9,layer10,layer11,layer12,layer13,layer14,layer15])
model.compile(loss='mean_absolute_error',optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history=model.fit(celsius_q,fahrenheit_a,epochs=1000,verbose=True)
print("Finished training the model")
```
